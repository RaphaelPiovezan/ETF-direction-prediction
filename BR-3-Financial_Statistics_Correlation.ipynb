## Volatility Clustering

```python
logreturns = logreturns.iloc[: , 1:]
```

```python
for i, tickers in enumerate(logreturns.columns):
  plt.subplot(len(logreturns.columns), 1, i + 1)
  plt.title(tickers)
  logreturns[tickers].plot(figsize=(20,60));
```
