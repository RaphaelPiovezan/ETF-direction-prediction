## Financial Data Statistics

```python
import pandas as pd
import pandas_datareader as pdr
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

## Covariance and Correlation

```python
tickers = ['BOVA11.SA','SPY'
]

df0 = pdr.DataReader(tickers, data_source='yahoo', start='2012-01-01', end='2022-01-25')['Close']
df0.to_csv('BOVAvsSPY.csv', index = True)  
```

```python
df0
```

```python
df0_returns = pd.DataFrame()
for tickers in df0.columns:
  df0_returns[tickers] = np.log(df0[tickers]).diff()
  
```

```python
import seaborn as sns
sns.pairplot(df0_returns);
plt.savefig('paiplot SPY-BOVA.PDF', format='PDF')
```

## Volatility Clustering
